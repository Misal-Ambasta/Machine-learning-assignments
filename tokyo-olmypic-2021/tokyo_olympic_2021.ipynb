```python
# # Rename column for consistency
medals_renamed = medals.rename(columns={"Team/NOC": "NOC"})

# Count athletes by NOC
athletes_medal = (
    athletes.groupby("NOC")["Name"]
    .count()
    .reset_index(name="AthleteCount")
    .sort_values("AthleteCount", ascending=False)
)

# Merge with medal data
athletes_medal = athletes_medal.merge(medals_renamed, on="NOC", how="left").head(10)

athletes_medal
```

## Comparisons between continents/regions

```python
noc_to_continent = {
    "United States of America": "Americas",
    "Canada": "Americas",
    "Brazil": "Americas",
    "Argentina": "Americas",

    "People's Republic of China": "Asia",
    "Japan": "Asia",
    "Republic of Korea": "Asia",
    "India": "Asia",
    "Iran": "Asia",

    "Great Britain": "Europe",
    "Germany": "Europe",
    "France": "Europe",
    "Italy": "Europe",
    "Netherlands": "Europe",
    "ROC": "Europe",   # Russian Olympic Committee
    "Spain": "Europe",

    "Australia": "Oceania",
    "New Zealand": "Oceania",

    "Kenya": "Africa",
    "South Africa": "Africa",
    "Egypt": "Africa",
    "Nigeria": "Africa",
}
```

```python

medals["Continent"] = medals["Team/NOC"].map(noc_to_continent)

medals
```

```python
nan_teams  = medals[medals["Continent"].isna()][["Rank", "Team/NOC"]]

nan_count = nan_teams["Team/NOC"].nunique()

nan_count, nan_teams.sort_values("Rank").reset_index(drop=True)
```
